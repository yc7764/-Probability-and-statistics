```python
cls = setup(data=df, target = '결과',session_id = SEED,
            numeric_features = numeric_features,
            ignore_features=['날짜','팀명'], categorical_imputation='mode',
            train_size=0.8,
            profile=True)
```

```python
cls
```

```python
best_models = compare_models(sort='Accuracy',n_select=3)
```

```python
best_models
```

```python
#모델 조합
blended = blend_models(estimator_list=best_models, fold=5)
```

```python
mymodels = tune_model(blended)
```

```python
tune2 = tune_model(mymodels)
```

```python
#최종 모델 학습
final_model = finalize_model(mymodels)
```

```python
#예측
pred = predict_model(final_model, data=df)
pred
```

```python
prediction = pred['Label']
prediction
```

```python
from pycaret.utils import check_metric
```
