```python
!pip install selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
pd.set_option('display.max_columns', None) # 데이터프레임 컬럼 display

import warnings
warnings.filterwarnings('ignore') # 경고 무시

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
# 반복문으로 각 url 출력
seasons = [f'0{num}' for num in range(14, 19)]
yymms = [
pd.date_range('2017-11', '2018-04', freq = 'M').strftime('%Y-%m').to_list(),
pd.date_range('2018-10', '2019-04', freq = 'M').strftime('%Y-%m').to_list(),
pd.date_range('2019-10', '2020-04', freq = 'M').strftime('%Y-%m').to_list(),
pd.date_range('2020-10', '2021-05', freq = 'M').strftime('%Y-%m').to_list(),
pd.date_range('2021-10', '2022-03', freq = 'M').strftime('%Y-%m').to_list()
]
# print(seasons)
# print(yymm)

for season, dates in zip(seasons, yymms) :
    for date in dates :
        url = f'https://www.kovo.co.kr/game/v-league/11110_schedule_list.asp?season={season}&team=&yymm={date}&r_round='
        print(url)
```

```python
def get_result(driver) :
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find_all('table')

    team1 = pd.read_html(str(table))[0].reset_index(drop=False)
    
    date_selector = '#wrp_content > article.wrp_recentgame.wrp_result > table > tbody > tr > td:nth-child(3) > table > tbody > tr:nth-child(1) > td:nth-child(2)'
    date = soup.select_one(date_selector).get_text()
    
    return date
```

```python
def get_result(driver) :
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find_all('table')
    result = pd.read_html(str(table))[1]
    
    return result
```

```python
def c_frame(dtframe):
    df1 = dtframe.drop(2)
    df2 = pd.DataFrame(dtframe.iloc[2,1:]).T
    col_name = ['경기시간_1set', '경기시간_2set','경기시간_3set', '경기시간_4set', '경기시간_5set', '경기시간_total']
    df2 = df2.rename(columns={ df2.columns[i] : col_name[i] for i in range(6)})
    for idx in df2.columns:
        df1[idx] = df2.loc[2, idx]
    return df1
```

```python
driver = webdriver.Chrome()
url = "https://www.kovo.co.kr/game/v-league/11110_schedule_list.asp?season=014&team=&yymm=2017-10&r_round="
driver.get(url)
time.sleep(2)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
table = soup.find_all('table')
df = pd.read_html(str(table))[0]
idx = df[df['구분'] == '여자'].index.to_list()
idx = [i + 1 for i in idx]

dfs = []

try :
    x_path = f'//*[@id="type1"]/div/table/tbody/tr[{idx[0]}]/td[10]/a[3]' # 상세결과 버튼
    cursor = driver.find_element(By.XPATH,x_path)
except :
    x_path = f'//*[@id="type1"]/div/table/tbody/tr[{idx[0]}]/td[10]/a[2]' # 상세결과 버튼
    cursor = driver.find_element(By.XPATH,x_path)
cursor.click()
time.sleep(1)

temp_df = get_result(driver)

team_result = c_frame(temp_df)

team_result
```
